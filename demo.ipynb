# KMeans-GPU

```python
import torch
from kmeans_gpu import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed
np.random.seed(123)
```

```python
# Generate data
data_size, dims, num_clusters = 1000, 2, 3
x = np.random.randn(data_size, dims) / 6
x = torch.from_numpy(x).float()
```

```python
kmeans = KMeans(n_clusters=num_clusters, distance="euclidean")
```

```python
batch_x = x.unsqueeze(dim=0)
center_pts = kmeans(batch_x)
center_pts = center_pts.squeeze(dim=0)

print(x.shape, center_pts.shape)
```

```python
# plot
plt.scatter(x[:, 0], x[:, 1])
plt.scatter(center_pts[:, 0], center_pts[:, 1], c="white", alpha=0.6, edgecolors='black', linewidth=2)

plt.axis([-1, 1, -1, 1])
plt.tight_layout()
plt.show()
```
